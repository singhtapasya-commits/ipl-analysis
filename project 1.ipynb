VIRAT KOHLI VS ROHIT SHARMA ipl analysis

# 🏏 IPL Data Analysis: Virat Kohli vs Rohit Sharma

### 📌 Objective:
To analyze and compare the IPL performance of Virat Kohli and Rohit Sharma using data analysis and visualization.

### 📊 Tools Used:
- Python
- Pandas
- Matplotlib


## 📂 Dataset
- Source: IPL Data (2008–2017)


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sharma-kohli.csv')
```

```python
print(df.head())
```

## 📂 Understanding the Dataset
We will check structure, columns, and missing values.


```python
df.info()
df.describe()
```

TOTAL RUNS

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
rohit_total = df['RG Sharma'].sum()
kohli_total = df['V Kohli'].sum()

print("Rohit Sharma Total Runs:", rohit_total)
print("Virat Kohli Total Runs:", kohli_total)
```

LINE PLOT

```python
plt.figure(figsize=(12,4))
plt.plot(df['index'],df['RG Sharma'],label='R Sharma',marker='s')
plt.plot(df['index'],df['V Kohli'],label='V Kohli',color='r',marker='o')
plt.title('Sharma VS Kohli performance')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend(loc='best')   #to use labels and colrs without it we cannot use this
plt.grid(True)
plt.show()
```

AVERAGE

```python
print("Rohit Average:", df['RG Sharma'].mean())
print("Kohli Average:", df['V Kohli'].mean())
```

```python
players = ['Rohit Sharma', 'Virat Kohli']
runs = [rohit_total, kohli_total]

plt.bar(players, runs,color=['pink','yellow'])

plt.title('Total Runs Comparison')
plt.ylabel('Runs')

plt.show()
```

```python
rohit_best = df.loc[df['RG Sharma'].idxmax()]
kohli_best = df.loc[df['V Kohli'].idxmax()]

print("Rohit Best Season:")
print(rohit_best)

print("\nKohli Best Season:")
print(kohli_best)
```

## 🔍 Insights

- Virat Kohli has higher peak performance (973 runs in a season)
- Rohit Sharma shows more balanced performance across seasons
- Kohli dominates in high-scoring seasons
- Rohit is more consistent but with fewer extreme peaks


PIE CHART

```python
plt.figure(figsize=(6,6))

plt.pie(runs, labels=players, autopct='%1.1f%%',
        colors=['#74b9ff', '#ff7675'],
        startangle=90)

plt.title('Run Contribution Share')

plt.show()

```

```python
plt.figure(figsize=(12,5))

plt.plot(df['index'], df['RG Sharma'], marker='o', label='Rohit Sharma')
plt.plot(df['index'], df['V Kohli'], marker='o', label='Virat Kohli')

# Highlight max points
plt.scatter(df['index'][df['RG Sharma'].idxmax()],
            df['RG Sharma'].max(), color='blue', s=100)

plt.scatter(df['index'][df['V Kohli'].idxmax()],
            df['V Kohli'].max(), color='red', s=100)

plt.title('Performance with Peak Highlight')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend()
plt.grid(True)

plt.show()
```

## ✅ Conclusion

Both players are exceptional in IPL history.

- Virat Kohli excels in high-scoring seasons and has record-breaking performances.
- Rohit Sharma shows consistent contribution over the years.

This analysis highlights different playing styles:
Consistency vs Peak Performance.
